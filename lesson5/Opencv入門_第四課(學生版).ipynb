```python
# 例子一：找出 lena.jpg 中的人臉和眼部

import cv2

img = cv2.imread('lena.jpg')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# 載入haar眼部分類器
eye_cascade = cv2.CascadeClassifier('haarcascade_eye_tree_eyeglasses.xml')
# 載入haar臉部分類器
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# 使用haar臉部分類器
faces = face_cascade.detectMultiScale(img_gray, 1.1, 3)

for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,0,255), 2)
    
    # 為提高眼部檢測的準確度和計算速度，eyes_area只有面部圖像
    eyes_area = img[y:y+h, x:x+w]
    eyes_area_gray = img_gray[y:y+h, x:x+w]
    # 使用haar眼部分類器
    eyes = eye_cascade.detectMultiScale(eyes_area, 1.1, 3)
    for (eyes_x, eyes_y, eyes_w, eyes_h) in eyes:
        cv2.rectangle(img, (x+eyes_x,y+eyes_y), (x+eyes_x+eyes_w,y+eyes_y+eyes_h), (0,255,0), 2)
    
    text = 'Faces & Eyes Detection'
    font = cv2.FONT_HERSHEY_SIMPLEX
    size = 0.4
    img = cv2.putText(img, text, (x,y+h+10), font, size, (255,255,255), 1, cv2.LINE_AA)

# cv2.imwrite('sample_4_1.jpg',img)
cv2.imshow('sample_4_1',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 例子二：用「白色圓點」標記出「girls.jpg」中所有女生的眼部

import cv2

img = cv2.imread('girls.jpg')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye_tree_eyeglasses.xml')
faces = face_cascade.detectMultiScale(img_gray, 1.1, 3)

for (x,y,w,h) in faces:
    eyes_area = img[y:y+h, x:x+w]
    eyes_area_gray = img_gray[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(eyes_area, 1.1, 3)
    for (eyes_x, eyes_y, eyes_w, eyes_h) in eyes:
        cv2.circle(img, ((x+eyes_x+eyes_w//2),(y+eyes_y+eyes_h//2)), 1, (0,255,255), 3)
        
# cv2.imwrite('sample_4_2.jpg',img)
cv2.imshow('sample_4_2',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
