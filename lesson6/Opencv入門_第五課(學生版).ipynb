```python
# 例子一：讀取並顯示正常速度的 talk.mp4 全彩影片

import cv2  

# 讀取視頻
cap = cv2.VideoCapture('talk.mp4')
# 首先讀取第一幀畫面
ret, frame = cap.read()

# 當第一幀畫面讀取成功則進入循環
while (ret == True):
    # 顯示當前畫面
    cv2.imshow('video', frame)
    # 讀取下一幀畫面，若無法讀取則退出循環
    ret, frame = cap.read()
    # 視頻刷新速率(數字越小 -> 刷新速度越快)、按鍵事件
    k = cv2.waitKey(25)
    if (k & 0xff == ord('q')):  
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# 例子二：讀取攝像頭畫面並顯示正常速度的全彩畫面

import cv2

# 讀取攝像頭(0為預設攝像頭、1/2/3...為外置攝像頭)
cap = cv2.VideoCapture(0)
# 首先讀取第一幀畫面
ret, frame = cap.read()

# 當第一幀畫面截取成功則進入循環
while (ret == True):
    # 0/1/-1 水平上下翻轉影像
    frame = cv2.flip(frame, 1)
    # 顯示當前畫面
    cv2.imshow('camera', frame)
    # 讀取下一幀畫面，讀取則退出循環
    ret, frame = cap.read()
    # 視頻刷新速率(數字越小 -> 刷新速度越快)、按鍵事件
    k = cv2.waitKey(20)
    if (k & 0xff == ord('q')):
        break

cap.release()
cv2.destroyAllWindows()

```

```python
# 例子三：讀取攝像頭畫面並錄製成 .mp4 全彩影片檔案

import cv2

cap = cv2.VideoCapture(0)

# 設置錄製影片的命名、格式、速度、畫面尺寸
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('sample_5_3.mp4', fourcc, 20.0, (640,  480))

ret, frame = cap.read()

while (ret == True):
    # 0/1/-1 水平上下翻轉影像
    frame = cv2.flip(frame, 1)
    
    # 將當前畫面錄製
    out.write(frame)
    
    cv2.imshow('frame', frame)
    ret, frame = cap.read()
    k = cv2.waitKey(20)
    if (k & 0xff == ord('q')):
        break

# 結束錄製
out.release()

cap.release()
cv2.destroyAllWindows()
```
